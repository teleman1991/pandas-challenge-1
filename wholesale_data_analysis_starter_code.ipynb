## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
subtotal = df["unit_price"]*df['qty']
df["line_subtotal"] = subtotal
df[["unit_price", "qty", "line_subtotal"]].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
total_weight = df["unit_weight"]*df['qty']
df["total_weight"] = total_weight

shipping = df.apply(lambda x: x['total_weight'] * 7 \
                    if x['total_weight'] > 50 else x['total_weight'] * 10, axis=1)

shipping_price = df["shipping_price"] = shipping



df[["unit_price", "unit_weight", "qty", "total_weight", "shipping_price"]].head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
tax = 0.0925
total_price = df['line_subtotal'] +df['shipping_price']
df["total_price"] = total_price
df["total_price"] *= 1 + tax

df[['line_subtotal', 'shipping_price', 'total_price']].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
line_cost = (df['unit_cost']*df['qty']) + df['shipping_price']
df['line_cost'] = line_cost
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
line_profit = df['total_price'] - df['line_cost']
df['line_profit'] = line_profit

df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
# Calculate the total cost for the top 5 clients

top_clients_cost1 = df[df['client_id'].isin(top_client_list_result.index)].groupby('client_id').apply(lambda x: (x['qty'] * x['unit_cost']).sum())

# Sort the result in descending order
top_clients_cost = top_clients_cost1.sort_values(ascending=False)

# Print the formatted output
for client_id, total_cost in top_clients_cost.items():
    print(f"{client_id}: ${total_cost:,.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_clients_ids = df['client_id'].value_counts().head(5).index

# Calculate total units purchased by each of the top 5 clients
top_clients_units = df[df['client_id'].isin(top_5_clients_ids)].groupby('client_id')['qty'].sum()

# Calculate total shipping price from each of the top 5 clients
top_clients_shipping = df[df['client_id'].isin(top_5_clients_ids)].groupby('client_id').apply(lambda x: (x['shipping_price']).sum())

# Calculate total revenue from each of the top 5 clients
top_clients_revenue = df[df['client_id'].isin(top_5_clients_ids)].groupby('client_id').apply(lambda x: (x['qty'] * x['unit_price']).sum())

# Calculate total cost from each of the top 5 clients
top_clients_cost = df[df['client_id'].isin(top_5_clients_ids)].groupby('client_id').apply(lambda x: (x['qty'] * x['unit_cost']).sum())

# Calculate total profit from each of the top 5 clients
top_clients_profit = top_clients_revenue - top_clients_cost

# Assign names to the Series
top_clients_shipping.name = 'Total Shipping Price (millions)'
top_clients_revenue.name = 'Total Revenue (millions)'
top_clients_profit.name = 'Total Profit (millions)'

# Merge 'first' and 'last' columns with 'client_id' from the main data
top_clients_info = df[['client_id', 'first', 'last']].drop_duplicates().set_index('client_id')

# Merge with top clients' units purchased
top_clients_info = top_clients_info.merge(top_clients_units, how='inner', left_index=True, right_index=True)

# Merge with top clients' total shipping price
top_clients_info = top_clients_info.merge(top_clients_shipping, how='inner', left_index=True, right_index=True)

# Merge with top clients' total revenue
top_clients_info = top_clients_info.merge(top_clients_revenue, how='inner', left_index=True, right_index=True)

# Merge with top clients' total profit
top_clients_info = top_clients_info.merge(top_clients_profit, how='inner', left_index=True, right_index=True)

# Create the summary DataFrame
summary_df = top_clients_info.reset_index()

# Sort the summary DataFrame by total profit
summary_df = summary_df.sort_values(by='Total Profit (millions)', ascending=False)

summary_df
```

```python
# Define a function that converts a dollar amount to millions.
def convert_to_millions(summary_df, column_name):
    """
    Converts the values of the specified column in the DataFrame to millions.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the column to be converted.
    column_name (str): The name of the column to convert.

    Returns:
    pd.DataFrame: The DataFrame with the specified column values converted to millions.
    """
    summary_df[column_name] = summary_df[column_name] / 1000000
    return summary_df
# List of columns to convert to millions # Apply the currency_format_millions function to only the money columns. 
columns_to_convert = ['Total Shipping Price (millions)', 'Total Revenue (millions)', 'Total Profit (millions)']
# Convert the specified columns to millions # Apply the currency_format_millions function to only the money columns. 
summary_df_millions = convert_to_millions(summary_df, columns_to_convert)
# Convert the specified columns to millions
summary_df_millions


```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
Profit_sort = summary_df_millions.sort_values(by='Total Profit (millions)', ascending=False)
Profit_sort
```
